### Embeddings de sentença e clusters

```python
!pip install -q gdown

import pandas as pd
import gdown

file_id = '1ZyTlST3HpdvW0db_Yl8Im8EL9YoLl3GU'
output_path = 'df.csv'

gdown.download(id=file_id, output=output_path, quiet=False)

df = pd.read_csv(output_path, delimiter='|', quotechar='"')
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "sentence-transformers/all-mpnet-base-v2"
)

emails = df["reprocessed_text"].astype(str).tolist()

embeddings2 = model.encode(
    emails,
    show_progress_bar=True
)

df["transformer_embeddings2"] = pd.Series(list(embeddings2))
```

```python
import umap
import hdbscan

reducer = umap.UMAP(
    n_components=20,
    n_neighbors=15,
    min_dist=0
)

X_red = reducer.fit_transform(embeddings2)

clusterer = hdbscan.HDBSCAN(
    min_cluster_size=10
)

df['clusters2'] = clusterer.fit_predict(X_red)
```

```python
# @title Gráfico de Clusters
import umap
import plotly.express as px
import pandas as pd

umap_reducer = umap.UMAP(n_components=3, random_state=42, n_jobs=1)

umap_components = umap_reducer.fit_transform(embeddings_array)

umap_df = pd.DataFrame(data=umap_components, columns=['UMAP1', 'UMAP2', 'UMAP3'])

if 'clusters' in df_sample_filt.columns and len(df_sample_filt['clusters'].unique()) > 1:
    umap_df['Cluster'] = df_sample_filt['clusters'].astype(str).values
    fig = px.scatter_3d(umap_df, x='UMAP1', y='UMAP2', z='UMAP3', color='Cluster',
                        title='UMAP dos Embeddings',
                        labels={'UMAP1': 'Component 1', 'UMAP2': 'Component 2', 'UMAP3': 'Component 3'})
else:
    fig = px.scatter_3d(umap_df, x='UMAP1', y='UMAP2', z='UMAP3',
                        title='UMAP dos Embeddings',
                        labels={'UMAP1': 'UMAP Component 1', 'UMAP2': 'UMAP Component 2', 'UMAP3': 'UMAP Component 3'})

fig.show()
```

### Avaliação NER

```python
import json
from collections import Counter

company_entities = Counter()
person_entities = Counter()
location_entities = Counter()
travel_location_entities = Counter()


for ner_entry in df['NER']:
    if pd.isna(ner_entry) or ner_entry == '{}':
        continue
    try:
        ner_data = json.loads(ner_entry)
        for entity_type, entities_list in ner_data.items():
            for entity in entities_list:
                if entity_type == 'company':
                    company_entities[entity.lower()] += 1
                elif entity_type == 'person':
                    person_entities[entity.lower()] += 1
                elif entity_type == 'location':
                    location_entities[entity.lower()] += 1
                elif entity_type == 'travel_location':
                    travel_location_entities[entity.lower()] += 1
    except json.JSONDecodeError:
        #print(f"Could not decode JSON: {ner_entry}")
        continue

print("\nTop most common Company Entities:")
for entity, count in company_entities.most_common(20):
    print(f"- {entity}: {count}")

print("\nTop most common Person Entities:")
for entity, count in person_entities.most_common(20):
    print(f"- {entity}: {count}")

print("\nTop most common Location Entities:")
for entity, count in location_entities.most_common(20):
    print(f"- {entity}: {count}")

print("\nTop most common Travel Location Entities:")
for entity, count in travel_location_entities.most_common(20):
    print(f"- {entity}: {count}")
```
