```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

file_list = []

pages = 40 #quantidade de páginas para ler. Costuma começar a baixar arquivos corrompidos entre 40 e 60 páginas
for i in range(pages):
    file_links = wd.driver.find_elements(By.CSS_SELECTOR, 'div.item-list > ul > li > div.views-field.views-field-title > span.field-content > a')
    file_list = file_list + [item.get_attribute('innerHTML') for item in file_links]
    time.sleep(1)
    try:
        next_page = wd.driver.find_element(By.CSS_SELECTOR, 'nav.usa-pagination > ul > li.usa-pagination__arrow:nth-last-child(2) > a.usa-pagination__link.usa-pagination__next-page')
        wd.driver.execute_script('document.querySelector("nav.usa-pagination > ul > li.usa-pagination__arrow:nth-last-child(2) > a.usa-pagination__link.usa-pagination__next-page").click()')
    except NoSuchElementException:
        break
```
